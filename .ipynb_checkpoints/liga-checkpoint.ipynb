# 5. Importar algoritmos de machine learnign

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#PCA

scaled_features = StandardScaler().fit_transform(jugadores_df.drop('Posicion', axis=1))
players_df_scaled = pd.DataFrame(scaled_features, index=jugadores_df.drop('Posicion', axis=1).index, columns=jugadores_df.drop('Posicion', axis=1).columns)
features_pca = PCA(n_components = 2).fit_transform(players_df_scaled)
principal_df = pd.DataFrame(features_pca, index=players_df_scaled.index, columns=["PC1", "PC2"])
```

```python

wcss=[]
for i in range(1,14):
    kmeans=KMeans(n_clusters=i,max_iter=300)
    kmeans.fit(principal_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,14), wcss)
plt.title("Codo de Jambu")
plt.xlabel("Numero de Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
### CÁLCULO DE KMEANS CON EL NÚMERO DE CLUSTER ÓPTIMOS

kmeans=KMeans(n_clusters=5)
y_kmeans = kmeans.fit_predict(principal_df)
kmeans.fit(principal_df)
centroides=kmeans.cluster_centers_
centroides
```

```python
### VEMOS LAS ETIQUETAS
kmeans.labels_
```

```python
### CREO LA COLUMNA KMEANS_CLUSTERS
jugadores_df['Kmeans_Clusters'] = kmeans.labels_
```

```python
### VEMOS LOS ÚLTIMOS 5 REGISTROS
jugadores_df.head(50)
```

```python
# Dibujar Clusters
plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.set_facecolor('#E8E8F1')
plt.xlabel('CP1', fontsize=18)
plt.ylabel('CP2', fontsize=18)

pos_colors = {
    'GoalKeeper': 'black', 
    'Defender': 'red',
    'Midfielder': 'yellow',
    'Attacker': 'green',
    'Forward': 'green'
}
plt.scatter(principal_df['PC1'], principal_df['PC2'], marker='o', c=jugadores_df['Posicion'].apply(lambda x: pos_colors[x]), cmap="Paired", s=50, alpha=0.5)

# Dibujar Centros
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], marker='o', c='#999999', s=500, alpha=0.7)

#texts = [plt.text(principal_df['PC1'][name], principal_df['PC2'][name], name) for name in principal_df['PC1'].index]
#adjust_text(texts, arrowprops=dict(arrowstyle='->', color='#999999'))

def encircle(x,y, ax=None, **kw):
    if not ax: ax=plt.gca()
    p = np.c_[x,y]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices,:], **kw)
    ax.add_patch(poly)


# Dibujar Poligonos Alrededor del vertice
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']
for i in y_kmeans:
    encircle(principal_df.loc[kmeans.labels_ == i, 'PC1'], principal_df.loc[kmeans.labels_ == i, 'PC2'], ec="k", fc=colors[i], alpha=0.02, linewidth=0)
```
